```python
# Import Splinter and BeautifulSoup
from splinter import Browser
from bs4 import BeautifulSoup
```

```python
browser = Browser('chrome')
```

### Step 1: Visit the Website

1. Use automated browsing to visit the [Mars news site](https://static.bc-edx.com/data/web/mars_news/index.html). Inspect the page to identify which elements to scrape.

      > **Hint** To identify which elements to scrape, you might want to inspect the page by using Chrome DevTools.

```python
# Visit the Mars news site
url = 'https://static.bc-edx.com/data/web/mars_facts/index.html'
browser.visit(url)
```

```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

### Step 2: Scrape the Website

Create a Beautiful Soup object and use it to extract text elements from the website.

```python
text_elements = [element.string for element in soup.find_all(string=True)]
```

```python
# Loop through and print all the text elements
for element in text_elements:
    print(element)
```

```python


# Extract titles and preview text and store them in dictionaries
article_list = []

# Find all the news article elements
articles = soup.find_all('div', class_='list_text')

for article in articles:
    title = article.find('div', class_='content_title').text.strip()
    preview = article.find('div', class_='article_teaser_body').text.strip()

    # Create a dictionary for each article
    article_dict = {'title': title, 'preview': preview}
    
    # Append the dictionary to the list
    article_list.append(article_dict)

# Print the list of dictionaries
for article_dict in article_list:
    print(article_dict)
```

```python
# Loop through the text elements
# Extract the title and preview text from the elements
# Store each title and preview pair in a dictionary
# Add the dictionary to the list
```

```python
# Initialize variables to store title and preview text
current_title = None
current_preview = None

# Loop through the elements with 'string' argument
for element in soup.find_all(string=True):
    # Check if the text is a title (you can modify this condition based on the actual HTML structure)
    if 'Title:' in element:
        current_title = element.replace('Title:', '').strip()
    # Check if the text is a preview (you can modify this condition based on the actual HTML structure)
    elif 'Text:' in element:
        current_preview = element.replace('Text:', '').strip()
    
    # If both title and preview have been found, create a dictionary and add it to the list
    if current_title and current_preview:
        article_dict = {'title': current_title, 'preview': current_preview}
        article_list.append(article_dict)
        
        # Reset variables for the next iteration
        current_title = None
        current_preview = None

# Print the list of dictionaries
for idx, article_dict in enumerate(article_list, start=1):
    print(f"Article {idx}:")
    print(f"Title: {article_dict['title']}")
    print(f"Preview: {article_dict['preview']}")
    print()
```
